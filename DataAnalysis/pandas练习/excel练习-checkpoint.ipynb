```python
"""创建、写入、保存一个xlsx"""
import xlwt
# 创建一个excel
wb = xlwt.Workbook()
# 选择工作薄
sh = wb.add_sheet("电影")
# 写入数据 到 单元格
sh.write(0,0,"复仇者联盟")
sh.write(2,6,"1000w")
# 保存excel
wb.save('./电影数据.xlsx')
```

```python
import xlrd
wb = xlrd.open_workbook("./电影数据.xlsx")
# 获取工作薄的个数和名称
print(wb.nsheets)
print(wb.sheet_names())
# 选择工作薄
sh1 = wb.sheet_by_index(0)
sh2 = wb.sheet_by_name("电影")
# 统计工作薄里面有多少行和列
print(f"行个数:{sh2.nrows}")
print(f"列个数:{sh2.ncols}")
# 获取单元格的值
print(f"第一行第一列的值:{sh2.cell_value(0,1)}")
print(f"第一行第一列的值:{sh2.cell(0,1).value}")
print(f"第一行第一列的值:{sh2.row(0)[1].value}")
# 获取整行或整列的数据
print(sh2.row_values(0))
print(sh2.col_values(0))
# 遍历所有数据
for r in range(sh1.nrows):
    for c in range(sh1.ncols):
        print(f"第{r}行，第{c}列的数据为{sh1.cell_value(r,c)}")      
```

```python
import xlrd
from xlutils.copy import copy

# 
read_book = xlrd.open_workbook("./电影数据.xlsx")
wb = copy(read_book)
# 选择工作薄
sh = wb.get_sheet(0)
# 添加数据
sh.write(4,0,"保卫家园")
sh.write(4,1,"2000")
sh.write(4,2,"六子")
# 添加工作薄
sh2 = wb.add_sheet("汇总数据")
count = 0
rs = read_book.sheet_by_index(0)
for i in range(1,rs.nrows):
    num = rs.cell_value(i, 1)
    print(num)
    count += int(num.replace("w", ""))
sh2.write(0,0,"总票房")
sh2.write(0,1,count)
wb.save("02_电影票房修改版.xlsx")
```

```python
import xlwt
path = "./03_test.xlsx"
wb = xlwt.Workbook(path)
sh = wb.add_sheet("电影")

# 一个Font代表一个组合样式
ft = xlwt.Font()
ft.colour_index = 2
ft.height = 11 * 20
ft.bold = True
ft.underline = True
ft.italic = True

# 用XFStyle对象让样式生效
style = xlwt.XFStyle()
style.font = ft
sh.write(6,0,"吕小布")
sh.write(6,1,"吕大部", style)

# 设置单元格高和宽
# 设置高度之前要启动设置高度模式
sh.row(3).height_mismatch = True
sh.row(3).height = 10 * 256
sh.col(3).width = 20 * 256

# 设置单元格排列方式
alg = xlwt.Alignment()
alg.horz = 2  # 左右对齐方式：1 左，2中，3右
alg.vert = 1  # 上下对齐方式：0 上，1中，2下
style2 = xlwt.XFStyle()
style2.alignment = alg
sh.write(3,3,"凋残",style2)

# 设置边框
border = xlwt.Borders()
border.left = 1
border.right = 1
border.top = 1
border.bottom = 1
border.left_colour = 1
border.right_colour = 2
border.bottom_colour = 3
border.bottom_colour = 4
style3 = xlwt.XFStyle()
style3.borders = border
sh.write(4,4,"边框",style3)

#设置背景颜色
pattern = xlwt.Pattern()
pattern.pattern = xlwt.Pattern.SOLID_PATTERN
pattern.pattern_fore_colour = 5
style4 = xlwt.XFStyle()
style4.pattern = pattern
sh.write(5,5,"周案例",style4)

# 简单设置，但是需要字符串特特殊格式
style5 = xlwt.easyxf('font: bold on,color_index 6;align:vert center, horiz center')
sh.write(6,6,"easygirl",style5)

wb.save(path)

```
